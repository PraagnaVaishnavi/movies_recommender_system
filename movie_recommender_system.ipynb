```python
import numpy as np
import pandas as pd
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
movies.shape


credits.head(1)

movies.head(1)

movies=movies.merge(credits,on='title')

movies.shape

#title genre id keywords overview cast crew
movies=movies[['id','title','overview','genres','keywords','cast','crew']]

movies.isnull().sum()
```

```python
#overview is having 3 null cells so drop it as only 3 rows
movies.dropna(inplace=True)
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj): #literal_eval will convert string object list into list of object
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    cnt = 0
    if pd.isna(obj):  # Handle NaN
        return L
    try:
        data = ast.literal_eval(obj)
    except (ValueError, SyntaxError):
        return L
    for i in data:
        if cnt != 3:
            cnt += 1
            L.append(i.get('name', ''))
        else:
            break
    return L

# Apply the function safely
movies['cast'] = movies['cast'].apply(convert3)
```

```python
import ast

def crewDirector(obj):
    L = []
    if pd.isna(obj):
        return L
    try:
        data = ast.literal_eval(obj)
    except (ValueError, SyntaxError):
        return L
    for i in data:
        if i.get('job', '').lower() == 'director':
            L.append(i.get('name', ''))
            break  # Only first director is enough
    return L

# Apply it
movies['crew'] = movies['crew'].apply(crewDirector)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

movies.head(1)

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['crew']
```

```python
movies['title']=movies['title'].str.lower()
movies
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x)) #converts list to integer
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))  # tenses of a word is considered as one ex dance dancing danced is stemmed as danc
    return  " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
print(similarity[0])
```

```python
def recommend(movie):
    movie=movie.lower()
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('John Carter')
```
